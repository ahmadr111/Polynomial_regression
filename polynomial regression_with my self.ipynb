```python
#split
x=df.drop('price',axis=1)
y=df['price']
```

```python
#train split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
print('x_train',x_train.shape)
print('y_train',y_train.shape)
print('x_test',x_test.shape)
print('y_test',y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

```python
std.fit(x_train)
```

```python
x_train_scaling=std.transform(x_train)
```

```python
x_test_scaling=std.transform(x_test)
```

```python
x_test_scaling
```

```python
x_train_scaling.mean().round(2)
```

```python
x_train_scaling.std().round(2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
poly.fit(x_train_scaling)
```

```python
x_train_poly=poly.transform(x_train_scaling)
x_test_poly=poly.transform(x_test_scaling)
```

```python
x_train_poly.shape
```

```python
x_test_poly.shape
```

```python
y_train.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train_poly,y_train)
```

```python
lr.score(x_test_poly,y_test)
```

```python
y_test_poly_predict=lr.predict(x_test_poly)
y_test_poly_predict
```

```python
y_test_poly_predict=lr.predict(x_test_poly)
y_test_poly_predict
```
